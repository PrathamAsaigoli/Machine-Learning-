```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from word2number import w2n
import warnings 
```

```python
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("hiring.csv")
df1 = df.copy()
df.info()
```

```python
df1
```

```python
df1.isnull().sum()
```

```python
test_score_median = df1['test_score(out of 10)'].median()
```

```python
#Converting words to numbers in the experience column
df1['experience'] = df1['experience'][2:8].apply(w2n.word_to_num)
```

```python
df1
```

```python
# Removing the NaN Values
# df['test_score(out of 10)']= df['test_score(out of 10)'].fillna(test_score_median)
df1= df1.fillna({'experience' : 0,'test_score(out of 10)' : test_score_median})
```

```python
df1
```

```python
df1.to_csv("df_refined.csv")
```

```python
X = df1[['experience','test_score(out of 10)','interview_score(out of 10)']]
y = df1[['salary($)']]
```

```python
reg = LinearRegression()
reg.fit(X,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[2,9,6]])
```
